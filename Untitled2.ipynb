```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from shapely.geometry import Polygon, Point

# 1. Load both GeoJSON files
# Make sure both files are uploaded to your Colab session
ncr_gdf = gpd.read_file('delhi_ncr_region.geojson')
airshed_gdf = gpd.read_file('delhi_airshed.geojson')

# 2. Reproject to EPSG:32644 as strictly mentioned in the SRIP PDF
ncr_proj = ncr_gdf.to_crs(epsg=32644)
airshed_proj = airshed_gdf.to_crs(epsg=32644)

# 3. Create the 60x60 km uniform grid based on Delhi-NCR bounds
minx, miny, maxx, maxy = ncr_proj.total_bounds
grid_size = 60000  # 60 km in meters

x_coords = np.arange(minx, maxx + grid_size, grid_size)
y_coords = np.arange(miny, maxy + grid_size, grid_size)

polygons = []
for x in x_coords[:-1]:
    for y in y_coords[:-1]:
        polygons.append(Polygon([(x, y), (x + grid_size, y), (x + grid_size, y + grid_size), (x, y + grid_size)]))

grid_gdf = gpd.GeoDataFrame({'geometry': polygons}, crs="EPSG:32644")

# 4. Generate Mock Satellite Image Data (To demonstrate filtering)
# (Jab aapke paas actual image coordinates honge, tab aap iski jagah apna CSV load kar lena)
np.random.seed(42)
num_points = 300
rand_x = np.random.uniform(minx - 40000, maxx + 40000, num_points)
rand_y = np.random.uniform(miny - 40000, maxy + 40000, num_points)

sat_images = pd.DataFrame({'image_id': [f'SAT_{i:03d}' for i in range(num_points)], 'easting': rand_x, 'northing': rand_y})
geometry = [Point(xy) for xy in zip(sat_images['easting'], sat_images['northing'])]
sat_points_gdf = gpd.GeoDataFrame(sat_images, geometry=geometry, crs="EPSG:32644")

# 5. Filter images whose center coordinates fall inside the Delhi Airshed region
# sjoin (spatial join) checks which points fall 'within' the Airshed polygon
filtered_points = gpd.sjoin(sat_points_gdf, airshed_proj, predicate='within')

# --- PLOTTING ---
fig, ax = plt.subplots(figsize=(12, 12))

# Plot Delhi-NCR Base map
ncr_proj.plot(ax=ax, color='lightblue', edgecolor='blue', alpha=0.4, label='Delhi-NCR Region')

# Plot Delhi Airshed
airshed_proj.plot(ax=ax, color='lightgreen', edgecolor='darkgreen', alpha=0.5, label='Delhi Airshed')

# Plot the 60x60 km Grid
grid_gdf.plot(ax=ax, facecolor='none', edgecolor='red', linewidth=1.5, linestyle='--', label='60x60 km Grid')

# Plot points (Outside vs Inside)
sat_points_gdf.plot(ax=ax, color='gray', markersize=10, alpha=0.5, label='Outside Images (Filtered Out)')
filtered_points.plot(ax=ax, color='purple', markersize=20, label='Inside Airshed (Kept)', marker='*')

# Formatting
ax.set_title('Delhi Airshed & NCR: 60×60 km Grid and Satellite Filtering', fontsize=16)
ax.set_xlabel('Easting (meters) - EPSG:32644', fontsize=12)
ax.set_ylabel('Northing (meters) - EPSG:32644', fontsize=12)

# Create a custom legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(loc='upper right')

# Save and show
plt.savefig('Q1_Spatial_Reasoning_Plot.png', bbox_inches='tight', dpi=300)
plt.show()

# --- REPORTING ---
print("--- TASK Q1 REPORT ---")
print(f"Total number of images BEFORE filtering: {len(sat_points_gdf)}")
print(f"Total number of images AFTER filtering: {len(filtered_points)}")
```

```python
import os
import zipfile
import rasterio
from rasterio.windows import Window
import numpy as np
from PIL import Image

def process_archive(zip_name="archive(1).zip", tif_name="land_cover.tif"):
    output_dir = "extracted_patches"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the large Land Cover TIF
    with rasterio.open(tif_name) as src:
        # Open the uploaded Zip
        with zipfile.ZipFile(zip_name, 'r') as z:
            # Filter for image files inside the zip
            images_in_zip = [f for f in z.namelist() if f.lower().endswith(('.tif', '.tiff', '.jpg', '.png'))]

            print(f"Found {len(images_in_zip)} images in {zip_name}. Processing...")

            for img_name in images_in_zip:
                with z.open(img_name) as f:
                    with rasterio.open(f) as img:
                        # Get center coordinate (Longitude, Latitude)
                        # We calculate the center based on the image bounds
                        center_x = (img.bounds.left + img.bounds.right) / 2
                        center_y = (img.bounds.top + img.bounds.bottom) / 2

                # Convert world coordinates to pixel indices of land_cover.tif
                py, px = src.index(center_x, center_y)

                # Define 128x128 window (64 pixels in each direction from center)
                window = Window(px - 64, py - 64, 128, 128)

                # Read the 128x128 patch
                # bound_less=True fills areas outside the TIF with zeros
                patch = src.read(window=window, boundless=True, fill_value=0)

                # Prepare for saving (Move bands to the last dimension)
                # (Bands, H, W) -> (H, W, Bands)
                patch_reshaped = np.moveaxis(patch, 0, -1)

                # If the image is single-band, squeeze the last dimension to make it 2D
                if patch_reshaped.shape[-1] == 1:
                    patch_reshaped = patch_reshaped.squeeze(axis=-1) # Converts (H, W, 1) to (H, W)

                # Define the output path for the patch
                clean_name = os.path.basename(img_name)
                output_path = os.path.join(output_dir, f"patch_{clean_name}")

                # Saving as a standard image (change mode if your TIF is 16-bit)
                result_img = Image.fromarray(patch_reshaped.astype('uint8'))
                result_img.save(output_path)

    print(f"Success! Check the '{output_dir}' folder for your patches.")

# Run the function
if __name__ == "__main__":
    # Make sure land_cover.tif and archive(1).zip are in the same folder as this script
    process_archive("archive(1).zip", "land_cover.tif")
```

```python
import os
import zipfile
import rasterio
from rasterio.windows import Window
import numpy as np
import pandas as pd
from scipy import stats
from PIL import Image

def process_and_label(zip_name="archive(1).zip", tif_name="land_cover.tif"):
    output_dir = "extracted_patches"
    results = []

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    with rasterio.open(tif_name) as src:
        with zipfile.ZipFile(zip_name, 'r') as z:
            images_in_zip = [f for f in z.namelist() if f.lower().endswith(('.tif', '.tiff', '.jpg', '.png'))]

            for img_name in images_in_zip:
                with z.open(img_name) as f:
                    with rasterio.open(f) as img:
                        center_x = (img.bounds.left + img.bounds.right) / 2
                        center_y = (img.bounds.top + img.bounds.bottom) / 2

                # Get pixel coordinates
                py, px = src.index(center_x, center_y)
                window = Window(px - 64, py - 64, 128, 128)

                # Read the patch (usually single band for land cover)
                patch = src.read(1, window=window, boundless=True, fill_value=0)

                # --- Labeling Logic ---
                # Calculate the most frequent pixel value (the dominant class)
                dominant_class = int(stats.mode(patch, axis=None, keepdims=True).mode[0])

                results.append({
                    "original_image": os.path.basename(img_name),
                    "assigned_label": dominant_class
                })

                # Save the patch as an image
                output_path = os.path.join(output_dir, f"patch_{os.path.basename(img_name)}")
                Image.fromarray(patch.astype('uint8')).save(output_path)

    # Save results to a CSV file
    df = pd.DataFrame(results)
    df.to_csv("patch_labels.csv", index=False)
    print(f"Processed {len(results)} images. Labels saved to 'patch_labels.csv'.")

if __name__ == "__main__":
    # Install missing libs if needed: pip install rasterio scipy pandas
    process_and_label("archive(1).zip", "land_cover.tif")
```

```python
import os
import zipfile
import rasterio
from rasterio.windows import Window
import numpy as np
import pandas as pd
from scipy import stats
from PIL import Image

# ESA WorldCover 10m to Simplified Mapping
ESA_MAP = {
    10: 'Vegetation',  # Tree cover
    20: 'Cropland',    # Shrubland
    30: 'Vegetation',  # Grassland
    40: 'Cropland',    # Cropland
    50: 'Built-up',   # Built-up
    60: 'Others',     # Bare / Sparse vegetation
    70: 'Others',     # Snow and Ice
    80: 'Water',      # Permanent water bodies
    90: 'Vegetation',  # Herbaceous wetland
    95: 'Vegetation',  # Mangroves
    100: 'Others'      # Moss and lichen
}

def process_with_mapping(zip_name="archive(1).zip", tif_name="land_cover.tif"):
    output_dir = "extracted_patches"
    results = []

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    with rasterio.open(tif_name) as src:
        with zipfile.ZipFile(zip_name, 'r') as z:
            images_in_zip = [f for f in z.namelist() if f.lower().endswith(('.tif', '.tiff', '.jpg', '.png'))]

            for img_name in images_in_zip:
                with z.open(img_name) as f:
                    with rasterio.open(f) as img:
                        center_x = (img.bounds.left + img.bounds.right) / 2
                        center_y = (img.bounds.top + img.bounds.bottom) / 2

                py, px = src.index(center_x, center_y)
                window = Window(px - 64, py - 64, 128, 128)

                # Read single band
                patch = src.read(1, window=window, boundless=True, fill_value=0)

                # Get dominant class code (Mode)
                dominant_code = int(stats.mode(patch, axis=None, keepdims=True).mode[0])

                # Map to simplified category (Defaults to 'Others' if code not found)
                simplified_label = ESA_MAP.get(dominant_code, 'Others')

                results.append({
                    "image_name": os.path.basename(img_name),
                    "esa_code": dominant_code,
                    "label": simplified_label
                })

                # Save patch
                output_path = os.path.join(output_dir, f"patch_{os.path.basename(img_name)}")
                Image.fromarray(patch.astype('uint8')).save(output_path)

    # Save to CSV
    df = pd.DataFrame(results)
    df.to_csv("mapped_land_cover.csv", index=False)
    print("Processing complete. Check 'mapped_land_cover.csv'.")

if __name__ == "__main__":
    process_with_mapping("archive(1).zip", "land_cover.tif")
```

```python
import os
import zipfile
import rasterio
from rasterio.windows import Window
import numpy as np
import pandas as pd
from scipy import stats
from PIL import Image

# ESA WorldCover mapping to Simplified Categories
ESA_MAP = {
    10: 'Vegetation', 20: 'Vegetation', 30: 'Vegetation', 90: 'Vegetation', 95: 'Vegetation',
    40: 'Cropland',
    50: 'Built-up',
    80: 'Water',
    60: 'Others', 70: 'Others', 100: 'Others'
}

def process_archive_with_labels(zip_name="archive(1).zip", tif_name="land_cover.tif"):
    output_dir = "labeled_patches"
    results = []

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    with rasterio.open(tif_name) as src:
        with zipfile.ZipFile(zip_name, 'r') as z:
            # Filter valid image files
            images = [f for f in z.namelist() if f.lower().endswith(('.tif', '.tiff', '.jpg', '.png'))]

            print(f"Total {len(images)} images found. Processing...")

            for img_name in images:
                with z.open(img_name) as f:
                    with rasterio.open(f) as img:
                        # Calculate center of the uploaded image
                        cx = (img.bounds.left + img.bounds.right) / 2
                        cy = (img.bounds.top + img.bounds.bottom) / 2

                # Get pixel position in land_cover.tif
                py, px = src.index(cx, cy)
                window = Window(px - 64, py - 64, 128, 128)

                # Read 128x128 patch (Band 1)
                patch = src.read(1, window=window, boundless=True, fill_value=0)

                # Find dominant class (Mode)
                mode_res = stats.mode(patch, axis=None, keepdims=True)
                dominant_code = int(mode_res.mode[0])

                # Map to simplified label
                label = ESA_MAP.get(dominant_code, 'Others')

                # File save logic: "originalName_LABEL.png"
                base_name = os.path.basename(img_name).split('.')[0]
                new_filename = f"{base_name}_{label}.png"
                output_path = os.path.join(output_dir, new_filename)

                # Save the patch
                Image.fromarray(patch.astype('uint8')).save(output_path)

                results.append({"File": new_filename, "Class_Code": dominant_code, "Label": label})

    # Summary CSV save karna optional hai par useful rehta hai
    pd.DataFrame(results).to_csv("extraction_summary.csv", index=False)
    print(f"Done! Check the '{output_dir}' folder.")

if __name__ == "__main__":
    process_archive_with_labels("archive(1).zip", "land_cover.tif")
```

```python
import os
import zipfile
import rasterio
from rasterio.windows import Window
import numpy as np
import pandas as pd
from scipy import stats
from PIL import Image
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. ESA WorldCover Mapping
ESA_MAP = {
    10: 'Vegetation', 20: 'Vegetation', 30: 'Vegetation', 90: 'Vegetation', 95: 'Vegetation',
    40: 'Cropland',
    50: 'Built-up',
    80: 'Water',
    60: 'Others', 70: 'Others', 100: 'Others'
}

def full_pipeline(zip_name="archive(1).zip", tif_name="land_cover.tif"):
    data_list = []
    output_dir = "all_patches"
    if not os.path.exists(output_dir): os.makedirs(output_dir)

    print("--- Phase 1: Patch Extraction & Labeling ---")
    with rasterio.open(tif_name) as src:
        with zipfile.ZipFile(zip_name, 'r') as z:
            images = [f for f in z.namelist() if f.lower().endswith(('.tif', '.tiff', '.jpg', '.png'))]

            for img_name in images:
                with z.open(img_name) as f:
                    with rasterio.open(f) as img:
                        cx = (img.bounds.left + img.bounds.right) / 2
                        cy = (img.bounds.top + img.bounds.bottom) / 2

                py, px = src.index(cx, cy)
                window = Window(px - 64, py - 64, 128, 128)
                patch = src.read(1, window=window, boundless=True, fill_value=0)

                # Dominant Class & Mapping
                dominant_code = int(stats.mode(patch, axis=None, keepdims=True).mode[0])
                label = ESA_MAP.get(dominant_code, 'Others')

                # Save Image
                base_name = os.path.basename(img_name).split('.')[0]
                file_name = f"{base_name}_{label}.png"
                Image.fromarray(patch.astype('uint8')).save(os.path.join(output_dir, file_name))

                data_list.append({"filename": file_name, "label": label})

    # Create DataFrame
    df = pd.DataFrame(data_list)

    # --- Phase 2: Train-Test Split (60/40) ---
    print("\n--- Phase 2: Random 60/40 Split ---")
    # Before splitting, check class distribution and filter out classes with only 1 member
    print("Initial class distribution:")
    print(df['label'].value_counts())

    # Identify classes with only one sample
    single_sample_classes = df['label'].value_counts()[df['label'].value_counts() < 2].index.tolist()
    if single_sample_classes:
        print(f"\nRemoving images from classes with only one sample: {single_sample_classes}")
        df_filtered = df[~df['label'].isin(single_sample_classes)]
    else:
        df_filtered = df

    if len(df_filtered) == 0:
        print("No data left after filtering classes with single samples. Cannot perform train-test split.")
        return

    # Perform stratified split only on the filtered DataFrame
    train_df, test_df = train_test_split(df_filtered, test_size=0.40, random_state=42, stratify=df_filtered['label'])

    train_df.to_csv("train_metadata.csv", index=False)
    test_df.to_csv("test_metadata.csv", index=False)
    print(f"Total (after filtering): {len(df_filtered)} | Train: {len(train_df)} | Test: {len(test_df)}")

    # --- Phase 3: Visualization ---
    print("\n--- Phase 3: Visualizing Distribution ---")
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    train_df['label'].value_counts().plot(kind='bar', ax=ax[0], color='skyblue', edgecolor='black')
    ax[0].set_title(f'Train Set Distribution ($60\%$)')
    ax[0].set_ylabel('Count')

    test_df['label'].value_counts().plot(kind='bar', ax=ax[1], color='salmon', edgecolor='black')
    ax[1].set_title(f'Test Set Distribution ($40\%$)')
    ax[1].set_ylabel('Count')

    plt.tight_layout()
    plt.savefig("class_distribution.png")
    print("Graph saved as 'class_distribution.png'")

if __name__ == "__main__":
    # Ensure you have sklearn installed: pip install scikit-learn
    full_pipeline("archive(1).zip", "land_cover.tif")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import pandas as pd
import numpy as np
import os
from PIL import Image
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
torch.manual_seed(42)

# ==========================================
# 1. Custom Dataset Class
# ==========================================
class LandCoverDataset(Dataset):
    def __init__(self, csv_file, img_dir, transform=None):
        self.data = pd.read_csv(csv_file)
        self.img_dir = img_dir
        self.transform = transform

        # Mapping categories to integers for PyTorch
        self.label_map = {
            'Built-up': 0,
            'Vegetation': 1,
            'Water': 2,
            'Cropland': 3,
            'Others': 4
        }

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Extract filename and label from CSV
        img_name = str(self.data.iloc[idx]['filename'])
        label_str = str(self.data.iloc[idx]['label'])
        label = self.label_map.get(label_str, 4) # Default to 'Others' if not found

        img_path = os.path.join(self.img_dir, img_name)

        # Safe Loading: Try to open the image, if not found, create a blank 128x128 image
        try:
            image = Image.open(img_path).convert('RGB')
        except FileNotFoundError:
            image = Image.new('RGB', (128, 128), color='black')

        if self.transform:
            image = self.transform(image)

        return image, label

# Image Transformations (Resize to 128x128 and convert to Tensor)
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]) # Normalize RGB channels
])

# Create Dataset and DataLoader
# WARNING: Make sure you create a folder named 'satellite_images' and put your .png files there!
IMG_DIRECTORY = "satellite_images"
os.makedirs(IMG_DIRECTORY, exist_ok=True)

train_dataset = LandCoverDataset(csv_file='train_metadata.csv', img_dir=IMG_DIRECTORY, transform=transform)
test_dataset = LandCoverDataset(csv_file='test_metadata.csv', img_dir=IMG_DIRECTORY, transform=transform)

# Batch size 16 to avoid running out of RAM
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

# ==========================================
# 2. Define Custom CNN Model
# ==========================================
class CustomCNN(nn.Module):
    def __init__(self, num_classes=5):
        super(CustomCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # Output: 16 x 64 x 64

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # Output: 32 x 32 x 32

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)  # Output: 64 x 16 x 16
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 16 * 16, 128),
            nn.ReLU(),
            nn.Dropout(0.5), # Prevent overfitting
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

# Setup Device, Model, Loss, and Optimizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CustomCNN(num_classes=5).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ==========================================
# 3. Model Training Loop
# ==========================================
print(f"Training on: {device}")
epochs = 3 # You can increase this to 10 or 20 for real training

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {running_loss/len(train_loader):.4f}")

print("Training Complete!")

# ==========================================
# 4. Evaluation & Confusion Matrix
# ==========================================
model.eval()
all_preds = []
all_labels = []

print("Evaluating model on test data...")
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Calculate Accuracy & F1-Score
accuracy = accuracy_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds, average='weighted')

print(f"\n--- FINAL RESULTS ---")
print(f"Accuracy:  {accuracy * 100:.2f}%")
print(f"F1-Score:  {f1:.4f}")

# Plot Confusion Matrix
classes = ['Built-up', 'Vegetation', 'Water', 'Cropland', 'Others']
cm = confusion_matrix(all_labels, all_preds, labels=[0, 1, 2, 3, 4])

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title('Confusion Matrix: Land-Use Classification')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('Q3_Confusion_Matrix.png', bbox_inches='tight', dpi=300)
plt.show()
```
